```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv('/content/Student_Performance.csv', sep=',')
df = df.dropna()
df.head()
```

```python
df1 = df

df1['Extracurricular Activities'] = df1.apply(lambda row: 1 if row['Extracurricular Activities']=='Yes' else 0, axis=1)

df1.head()
```

```python
model = LinearRegression()
x = df1[['Hours Studied',	'Previous Scores',	'Extracurricular Activities',	'Sleep Hours','Sample Question Papers Practiced']]
y = df1['Performance Index']

model.fit(x, y)
print(model.score(x, y))

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```
