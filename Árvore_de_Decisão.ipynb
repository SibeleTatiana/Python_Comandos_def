```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics

df = pd.read_csv('/full_data.csv')
df.head()
```

```python
df['gender'] = df.apply(lambda row: 1 if row['gender'] == 'Female' else 0, axis=1)
df.head()
```

```python
df['Residence_type'] = df.apply(lambda row: 1 if row['Residence_type'] == 'Rural' else 0, axis=1)
df.head()
```

```python
df['ever_married'] = df.apply(lambda row: 1 if row['ever_married'] == 'Yes' else 0, axis=1)
df.head()
```

```python
df['smoking_status'].value_counts()
```

```python
df['Unknown'] = df.apply(lambda row: 1 if row['smoking_status'] == 'Unknown' else 0, axis=1)
df['smokes'] = df.apply(lambda row: 1 if row['smoking_status'] == 'smokes' else 0, axis=1)
df['never_smoked'] = df.apply(lambda row: 1 if row['smoking_status'] == 'never smoked' else 0, axis=1)
df['formerly_smoked'] = df.apply(lambda row: 1 if row['smoking_status'] == 'formerly smoked' else 0, axis=1)
del df['smoking_status']

df.head()
```

```python
df['work_type'].value_counts()
```

```python
df['Private'] = df.apply(lambda row: 1 if row['work_type'] == 'Private' else 0, axis=1)
df['Self-employed	'] = df.apply(lambda row: 1 if row['work_type'] == 'Self-employed	' else 0, axis=1)
df['Govt_job'] = df.apply(lambda row: 1 if row['work_type'] == 'Govt_job' else 0, axis=1)
df['children'] = df.apply(lambda row: 1 if row['work_type'] == 'children' else 0, axis=1)
del df['work_type']
df

```

```python
df_stroked = df[df['stroke']==1].sample(248, random_state=42)
df_not_stroked = df[df['stroke']==0].sample(248, random_state=42)
df = pd.concat([df_stroked, df_not_stroked])
```

```python
x = df.drop('stroke', axis='columns')
y = df['stroke']
x_scaled = preprocessing.StandardScaler().fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=42)

# lbfgs, yes
# liblinear,no
# newton-cg,yes
# newton-cholesky,no
# sag,yes
# saga,yes

d = {'lbfgs':0, 'liblinear':0, 'newton-cg':0,  'newton-cholesky':0, 'sag':0, 'saga':0}

for k in d.keys():
  classifier = LogisticRegression(max_iter=100000, solver=k)
  classifier.fit(x_train, y_train)
  y_pred = classifier.predict(x_test)
  print(f"Accuracy {k}:", accuracy_score(y_test, y_pred))
  d[k] = accuracy_score(y_test, y_pred)

classifier = LogisticRegression(max_iter=100000, solver=max(d, key=d.get))
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
print(f"Chosen one ",max(d, key=d.get) ,":", accuracy_score(y_test, y_pred))
d[k] = accuracy_score(y_test, y_pred)
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier

```

```python
from sklearn import metrics

regressor = DecisionTreeClassifier(random_state=42, criterion='gini',max_depth=200, min_samples_leaf=50)
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)

# Accurary:  o nmero de previses corretas dividido pelo total de previses
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```
