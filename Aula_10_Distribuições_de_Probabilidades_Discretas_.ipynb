# **Distribuição Uniforme**

```python
# Definindo o número de faces do dado
num_faces = 6

# Calculando a probabilidade de cada face
probabilidade = 1 / num_faces

print("A probabilidade de cada face do dado de 6 faces é de", probabilidade)
```

```python
# Definindo o número de faces do dado
num_faces = 6

# Calculando a probabilidade de cada face
probabilidade = 1 / num_faces

# Criando os valores de x (faces do dado) e y (probabilidades)
faces = np.arange(1, num_faces+1)
probabilidades = np.full(num_faces, probabilidade)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.bar(faces, probabilidades, color='skyblue', alpha=0.7)
plt.title('Distribuição Uniforme de um Dado de 6 Faces')
plt.xlabel('Face do Dado')
plt.ylabel('Probabilidade')
plt.xticks(faces)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Horário de chegada dos ônibus em um sistema de transporte: Se os ônibus em um sistema de transporte público chegarem em intervalos regulares e cada intervalo de tempo tiver a mesma probabilidade de ocorrer, então o horário de chegada dos ônibus pode ser modelado como uma distribuição uniforme.

```python
# Definindo o intervalo de tempo entre os ônibus (em minutos)
intervalo_ônibus = 20

# Número total de intervalos de 20 minutos em um período de 1 hora (60 minutos)
n_intervalos = 60 // intervalo_ônibus

# Calculando a probabilidade de cada intervalo de tempo
probabilidade = 1 / n_intervalos

# Criando os valores de x (intervalos de tempo) e y (probabilidades)
intervalos_tempo = np.arange(intervalo_ônibus, 61, intervalo_ônibus)
probabilidades = np.full(n_intervalos, probabilidade)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(intervalos_tempo, probabilidades, width=5, color='skyblue', alpha=0.7)
plt.title('Distribuição Uniforme: Horário de Chegada dos Ônibus')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Probabilidade')
plt.xticks(intervalos_tempo)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Distribuição Binomial**


## **Uma empresa envia sms de propaganda para 10 clientes, a chance de 1 cliente acessar o link é de 50%. Qual a probabilidade de 3 clientes acessarem o link dos sms`s?**


```python
# Definindo os parâmetros da distribuição binomial
n = 10  # Número total de clientes
p = 0.5  # Probabilidade de um cliente acessar o link

# Número de clientes que queremos calcular a probabilidade de acessarem o link
k = 3

# Calculando a probabilidade usando a distribuição binomial
probabilidade = binom.pmf(k, n, p)

print("A probabilidade de 10 clientes acessarem o link é de\
aproximadamente {:.2f}%.".format(probabilidade * 100))
```

```python
# Definindo os parâmetros da distribuição binomial
n = 10   # Número total de clientes
p = 0.5  # Probabilidade de um cliente abrir o SMS

# Gerando os valores de x (quantidade de clientes que abrem o SMS)
x = np.arange(0, n+1)

# Calculando as probabilidades para cada valor de x
probabilidades = binom.pmf(x, n, p)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(x, probabilidades, color='skyblue', alpha=0.7)

# Adicionando os valores no topo de cada coluna
for i, prob in enumerate(probabilidades):
    plt.text(x[i], prob, f'{prob:.3f}', ha='center', va='bottom')

plt.title('Probabilidade de Clientes Abrirem o SMS')
plt.xlabel('Quantidade de Clientes que Abrem o SMS')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **A chance de acidente mensal em uma fábrica com 200 funcionários é de 3%, qual a probabilidade de 10 funcionários sofrerem acidente este mês?**

```python
# Definindo os parâmetros da distribuição binomial
n = 200   # Número total de funcionários
p = 0.03  # Probabilidade de um funcionário sofrer acidente
k = 10    # Número de funcionários que queremos calcular a probabilidade de sofrerem acidente

# Calculando a probabilidade usando a distribuição binomial
probabilidade = binom.pmf(k, n, p)

print("A probabilidade de 10 funcionários sofrerem acidente este mês é de aproximadamente {:.2f}%.".format(probabilidade * 100))
```

# **Distribuição de Poisson**

**Se uma empresa tem média mensal de 20 acidentes. Qual a probabilidade de ocorrer 7 acidentes?**

```python
import math

# Definindo a média mensal de acidentes (lambda)
lambd = 20

# Número de acidentes que queremos calcular a probabilidade
k = 7

# Calculando a probabilidade usando a fórmula da distribuição de Poisson
probabilidade = (math.exp(-lambd) * (lambd ** k)) / math.factorial(k)

print(f"A probabilidade de ocorrerem 7 acidentes \
é de aproximadamente {probabilidade:.4f}")
```

**Qual a probabilidade de ocorrer entre 15 a 17 acidentes? Mesma média.**

```python
import math

# Definindo a média mensal de acidentes (lambda)
lambd = 20

# Definindo os limites inferior e superior
limite_inferior = 15
limite_superior = 17

# Calculando a probabilidade de ocorrer até o limite superior
prob_acumulada_superior = 0
for k in range(limite_superior + 1):
    prob_acumulada_superior += (math.exp(-lambd) * (lambd ** k)) / math.factorial(k)

# Calculando a probabilidade de ocorrer até o limite inferior - 1
prob_acumulada_inferior = 0
for k in range(limite_inferior):
    prob_acumulada_inferior += (math.exp(-lambd) * (lambd ** k)) / math.factorial(k)

# Calculando a probabilidade entre os limites inferior e superior
probabilidade_entre = prob_acumulada_superior - prob_acumulada_inferior

print("A probabilidade de ocorrer entre 15 e 17 acidentes é de aproximadamente {:.4f}".format(probabilidade_entre))
```

**Diminuindo a quantidade de código escrito**

```python
# Definindo a média mensal de acidentes (lambda)
lambd = 20

# Definindo os limites inferiores e superiores
limite_inferior = 15
limite_superior = 17

# Calculando a probabilidade acumulada de 0 a limite_superior
prob_acumulada_superior = sum((math.exp(-lambd) * (lambd ** k)) / math.factorial(k) for k in range(limite_superior + 1))

# Calculando a probabilidade acumulada de 0 a limite_inferior - 1
prob_acumulada_inferior = sum((math.exp(-lambd) * (lambd ** k)) / math.factorial(k) for k in range(limite_inferior))

# Calculando a probabilidade entre limite_inferior e limite_superior
probabilidade_entre = prob_acumulada_superior - prob_acumulada_inferior

print("A probabilidade de ocorrer entre 15 e 17 acidentes é de aproximadamente {:.4f}".format(probabilidade_entre))
```

**errors='coerce'**: Este é um argumento opcional passado para pd.to_datetime(). Quando definido como 'coerce', isso significa que, se ocorrerem erros durante a conversão, os valores problemáticos serão definidos como NaT (Not a Time), ou seja, serão tratados como valores ausentes. Isso é útil quando há dados ruins ou ausentes na coluna que está sendo convertida.

```python
df['data_acidente'] = pd.to_datetime(df['data_acidente'], errors='coerce')
```

```python
#Poisson
n = 15
mu = 2 # indica com que frequência os eventos ocorrem em média em um determinado período
resultados = [i for i in range(n+1)]
dados_poisson = pd.DataFrame(resultados, columns=['n_sucessos'])
dados_poisson['probs'] = dados_poisson['n_sucessos'].apply(lambda x: poisson.pmf(k=x, mu=mu))
dados_poisson.plot(kind='bar',
                   x = 'n_sucessos',
                   y = 'probs',
                   legend = None)
```

```python
#Binomial
n = 5
p = 0.8
n_sucessos = [i for i in range(n+1)]
dados_binomial = pd.DataFrame(n_sucessos, columns=['n_sucessos'])
dados_binomial['probs'] = dados_binomial['n_sucessos'].apply(lambda x: binom.pmf(k=x, n = n, p = p))
dados_binomial['probs'] = dados_binomial['n_sucessos'].apply(lambda x: binom.pmf(k=x, n = n, p = p))
dados_binomial.plot(kind='bar',
                    x = 'n_sucessos',
                    y = 'probs',
                    legend = None)
```

```python
#Bernoulli
p = 0.2
resultados = np.arange(0, 2)
probabilidade = np.array([p, 1-p])
dados_bernoulli = pd.DataFrame(resultados, columns=['resultados'])
dados_bernoulli['probabilidade'] = probabilidade
dados_bernoulli.plot(kind = 'bar',
                     x = 'resultados',
                     y = 'probabilidade', width = 0.8)
```
