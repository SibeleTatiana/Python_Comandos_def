````python
Célula de texto <_ABMyDl3guFD>
# %% [markdown]


```
# This is formatted as code
```

**Teste-t com múltiplas amostras: ANOVA (stats.f_oneway)**
  **Determinação de diferença entre pares: Posthoc de teste-t (sp.posthoc_ttest)**

**Teste não pareado sem normalidade: Teste U de Mannwhitney (stats.mannwhitneyu)**

**Teste não pareado de múltiplas amostras sem normalidade: Teste de kruskal (stats.kruskal)**
  **Determinação de diferença entre pares: Posthoc de Mann Whitney (posthoc_mannwhitney)**

**Teste pareado, sem normalidade: Classificação sinalizada de Wilcoxon (wilcoxon)**

**Teste pareado, múltiplas amostras, sem normalidade: Qui-Quadrado de Friedman (stats.friedmanchisquare)**
  Determinação de diferença entre pares: Posthoc de Wilcoxon (sp.posthoc_wilcoxon)



Célula de texto <YXwltpL5TtU0>
# %% [markdown]

Teste de hipóteses

Se tem 1 amostra:
  Se eu conheço a média populacional: teste-z
  Se eu não conheço a média populacional: teste-t
Se tem 2 amostras:
  Se tem normalidade:
    Se tem variância igual:
      Se é amostra pareada
        **Uso teste-rel**
      Se não é amostra pareada:
        **Uso teste-ind**
    Se não tem variância igual:
      **Uso o teste de Welch**
  Se não tem normalidade:
    Se não é pareado:
      **Uso teste U**
    Se é pareado:
      **Uso Wilcoxon**
Se tem mais do que 2 amostras:
  Se tem normalidade:
    **Uso ANOVA**
  Se não tem normalidade:
    Se não é pareado:
      **Uso Kruskal**
    Se é pareado:
      **Uso Qui Quadrado de Friedman**






Célula de código <LrWoi283Mjb0>
# %% [code]
# Teste de Qui-Quadrado de Friedman
# Utilizado para múltiplas amostras, COM pareamento (AMOSTRAS PAREADAS) e sem normalidade
# Não preciso testar variância
# Para o Qui Quadrao de Friedman o tamanho das amostras tem que ser igual

# O que fazer se as amostras tem tamanho diferente?
#
# Exemplo:
# m1 = 20 observações
# m2 = 20 observações
# m3 = 30 observações
# aleatoriamente, retirar uma amostra de tamanho 20 do m3

# !pip install scikit_posthocs
import scipy.stats as stats
import scikit_posthocs as sp
import pandas as pd

# Nutricionista está fazendo um experimento.
# m1 - tem dieta diferente e sem exercício
# m2 - sem dieta, mas com exercício
# m3 - não tem dieta e nem exercício
# Avaliou o ganho de massa magra, depois de 3 meses
# existe diferente entre esses 3 grupos?

m1 = [794.1, 716.9, 993  , 724.7, 760.9, 908.2 , 659.3 , 690.8, 768.7, 717.3, 630.7 , 729.5, 714.1, 810.3, 583.5, 679.9 , 865.1]
m2 = [898.8, 881.2, 940.2, 966.2, 957.5, 1061.7, 1046.2, 980.4, 895.6, 919.7, 1074.1, 952.5, 796.3, 859.6, 871.1, 1047.5, 619]
m3 = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1 , 759.8 , 894.6, 867.6, 805.6, 765.4 , 800.3, 789.9, 875.3, 740  , 795.2 , 823.8]

# utilizar o Qui Quadrado de Friedman
# Para utilizar o Qui Quadrado de Friedman, é necessário que as amostras NÃO tenham normalidade

alpha = 0.05

estatistica_t, p_valor = stats.friedmanchisquare(m1, m2, m3)
# H0: todas as amostras são iguais
# H1: pelo menos uma amostra é diferente
if p_valor >= alpha:
  print("As amostras são iguais")
else:
  print("Pelo menos uma amostra é diferente")
  # Qual amostra é diferente???

  posthoc_df = sp.posthoc_wilcoxon([m1, m2, m3])
  nomes_grupos = ["Dieta", "Exerc", "Nada"]
  posthoc_df.columns = nomes_grupos
  posthoc_df.index = nomes_grupos
  print(posthoc_df)

Saída da execução
0KB
	Stream
		Pelo menos uma amostra é diferente
		          Dieta     Exerc      Nada
		Dieta  1.000000  0.001678  0.071411
		Exerc  0.001678  1.000000  0.002579
		Nada   0.071411  0.002579  1.000000

Célula de código <pQeh8ZF9Lg4s>
# %% [code]
# Teste de Wilcoxon

# Vou usar o teste Wilcoxon quando não houver normalidade, ou seja, se Shapiro falhou
# Funciona SOMENTE para comparação de amostras PAREADAS
# PAREADA: mesma origem, amostras que foram extraídas das mesmas condições. Lembrar da máquina que fábrica bolas!
# Não preciso testar Levene (Variância)

import scipy.stats as stats

m1 = [794.1, 716.9, 993, 724.7, 760.9, 908.2, 659.3, 690.8, 768.7, 717.3, 630.7, 729.5, 714.1, 810.3, 583.5, 679.9, 865.1]
m3 = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1, 759.8, 894.6, 867.6, 805.6, 765.4, 800.3, 789.9, 875.3, 740, 795.2, 823.8, 791.7, 948.5, 791.7, 948.5]

estatistica_t, p_valor = stats.wilcoxon(m1, m3)

# H0: diz que as amostras são iguais
# H1: as amostras são diferentes
if p_valor >= alpha:
  print("Pelo teste de Wilcoxon, as amostras são iguais")
else:
  print("Pelo teste de Wilcoxon, as amostras são diferentes")

Célula de código <6Orl8DM6JjhJ>
# %% [code]
# Teste de Kruskal
# Utilizado para múltiplas amostras, sem pareamento e sem normalidade

# !pip install scikit_posthocs
import scipy.stats as stats
import scikit_posthocs as sp
import pandas as pd

# Nutricionista está fazendo um experimento.
# m1 - tem dieta diferente e sem exercício
# m2 - sem dieta, mas com exercício
# m3 - não tem dieta e nem exercício
# Avaliou o ganho de massa magra, depois de 3 meses
# existe diferente entre esses 3 grupos?

m1 = [794.1, 716.9, 993, 724.7, 760.9, 908.2, 659.3, 690.8, 768.7, 717.3, 630.7, 729.5, 714.1, 810.3, 583.5, 679.9, 865.1]
m2 = [898.8, 881.2, 940.2, 966.2, 957.5, 1061.7, 1046.2, 980.4, 895.6, 919.7, 1074.1, 952.5, 796.3, 859.6, 871.1, 1047.5, 619]
m3 = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1, 759.8, 894.6, 867.6, 805.6, 765.4, 800.3, 789.9, 875.3, 740, 795.2, 823.8, 791.7, 948.5, 791.7, 948.5]

# utilizar o Kruskal
# Para utilizar o Kruskal, é necessário que as amostras NÃO tenham normalidade

alpha = 0.05

estatistica_t, p_valor = stats.kruskal(m1, m2, m3)
# H0: todas as amostras são iguais
# H1: pelo menos uma amostra é diferente
if p_valor >= alpha:
  print("As amostras são iguais")
else:
  print("Pelo menos uma amostra é diferente")
  # Qual amostra é diferente???

  posthoc_df = sp.posthoc_mannwhitney([m1, m2, m3])
  nomes_grupos = ["Dieta", "Exerc", "Nada"]
  posthoc_df.columns = nomes_grupos
  posthoc_df.index = nomes_grupos
  print(posthoc_df)

Saída da execução
0KB
	Stream
		Pelo menos uma amostra é diferente
		          Dieta     Exerc      Nada
		Dieta  1.000000  0.000228  0.039853
		Exerc  0.000228  1.000000  0.000660
		Nada   0.039853  0.000660  1.000000

Célula de código <izguvebVHyWg>
# %% [code]
# Teste U de Mann Whitney

# Vou usar o teste U quando não houver normalidade, ou seja, se Shapiro falhou, vou para o teste U
# Funciona para comparação de amostras NÃO PAREADAS

import scipy.stats as stats

m1 = [794.1, 716.9, 993, 724.7, 760.9, 908.2, 659.3, 690.8, 768.7, 717.3, 630.7, 729.5, 714.1, 810.3, 583.5, 679.9, 865.1]
m3 = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1, 759.8, 894.6, 867.6, 805.6, 765.4, 800.3, 789.9, 875.3, 740, 795.2, 823.8, 791.7, 948.5, 791.7, 948.5]

estatistica_t, p_valor = stats.mannwhitneyu(m1, m3)

# H0: diz que as amostras são iguais
# H1: as amostras são diferentes
if p_valor >= alpha:
  print("Pelo teste U, as amostras são iguais")
else:
  print("Pelo teste U, as amostras são diferentes")
Saída da execução
0KB
	Stream
		Pelo teste U, as amostras são diferentes

Célula de código <n0SvFM826y8w>
# %% [code]
# !pip install scikit_posthocs
import scipy.stats as stats
import scikit_posthocs as sp
import pandas as pd

# Nutricionista está fazendo um experimento.
# m1 - tem dieta diferente e sem exercício
# m2 - sem dieta, mas com exercício
# m3 - não tem dieta e nem exercício
# Avaliou o ganho de massa magra, depois de 3 meses
# existe diferente entre esses 3 grupos?

m1 = [794.1, 716.9, 993, 724.7, 760.9, 908.2, 659.3, 690.8, 768.7, 717.3, 630.7, 729.5, 714.1, 810.3, 583.5, 679.9, 865.1]
m2 = [898.8, 881.2, 940.2, 966.2, 957.5, 1061.7, 1046.2, 980.4, 895.6, 919.7, 1074.1, 952.5, 796.3, 859.6, 871.1, 1047.5, 619]
m3 = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1, 759.8, 894.6, 867.6, 805.6, 765.4, 800.3, 789.9, 875.3, 740, 795.2, 823.8, 791.7, 948.5, 791.7, 948.5]

# utilizar o ANOVA
# Para utilizar o ANOVA, é necessário que as amostras tenham normalidade E variância igual também

# verificar se tem normalidade
estatistica_t, p_valor_1 = stats.shapiro(m1)
estatistica_t, p_valor_2 = stats.shapiro(m2)
estatistica_t, p_valor_3 = stats.shapiro(m3)

alpha = 0.05

# Hipóteses Shapiro
# H0: tem normalidade
# H1: não tem normalidade

if p_valor_1 >= alpha and p_valor_2 >= alpha and p_valor_3 >= alpha:
  print("Amostras tem normalidade!")

  # Hipóteses Levene
  # H0: variância é igual
  # H1: variância é diferente
  estatistica_t, p_valor = stats.levene(m1, m2, m3)

  if p_valor >= alpha:
    print("Amostras tem variância igual")

    estatistica_t, p_valor = stats.f_oneway(m1, m2, m3)
    # H0: todas as amostras são iguais
    # H1: pelo menos uma amostra é diferente
    if p_valor >= alpha:
      print("As amostras são iguais")
    else:
      print("Pelo menos uma amostra é diferente")
      # Qual amostra é diferente???

      posthoc_df = sp.posthoc_ttest([m1, m2, m3])
      nomes_grupos = ["Dieta", "Exerc", "Nada"]
      posthoc_df.columns = nomes_grupos
      posthoc_df.index = nomes_grupos
      print(posthoc_df)
  else:
    print("Amostras tem variância diferente")
else:
  print("Minhas amostras não tem normalidade, não posso continuar com o anova")
Saída da execução
0KB
	Stream
		Amostras tem normalidade!
		Amostras tem variância igual
		Pelo menos uma amostra é diferente
		          Dieta     Exerc      Nada
		Dieta  1.000000  0.000030  0.050986
		Exerc  0.000030  1.000000  0.000828
		Nada   0.050986  0.000828  1.000000


````
