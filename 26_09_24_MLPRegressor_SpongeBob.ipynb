```python
df['_800'] = df['U.S. premiere time (ET)'].apply(lambda x: 1 if x == '8:00 pm' else 0)
df['_500'] = df['U.S. premiere time (ET)'].apply(lambda x: 1 if x == '5:00 pm' else 0)
df['_815'] = df['U.S. premiere time (ET)'].apply(lambda x: 1 if x == '8:15 pm' else 0)
df['_700'] = df['U.S. premiere time (ET)'].apply(lambda x: 1 if x == '7:00 pm' else 0)
df['_1000'] = df['U.S. premiere time (ET)'].apply(lambda x: 1 if x == '10:00 am' else 0)
df['_other'] = ((df['_800'] == 0) & (df['_500'] == 0) & (df['_700'] == 0) & (df['_815'] == 0) & (df['_1000'] == 0) ).astype(int)
df['Season №'] = df['Season №'].astype(int)
df = df.rename(columns={'Season №': 'season'})

df['fred'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'Fred Osmond' else 0)
df['jhon'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'John Trabbic' else 0)
df['Brian'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'Brian Morante' else 0)
df['Erik'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'Erik Wiese	' else 0)
df['Greenblatt'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'C.H. Greenblatt' else 0)
df['Kelly'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'Kelly Armstrong' else 0)
df['Caleb'] = df['Storyboard Artist(s)'].apply(lambda x: 1 if x == 'Caleb Meurer' else 0)
df['outros'] = ((df['fred'] == 0) & (df['jhon'] == 0) & (df['Brian'] == 0) & (df['Erik'] == 0) & (df['Greenblatt'] == 0) & (df['Kelly']== 0) & (df['Caleb']==0)).astype(int)
```

```python
df['U.S. viewers (millions)'] = pd.to_numeric(df['U.S. viewers (millions)'], errors='coerce')
df['U.S. viewers (millions)'].describe()
df = df.dropna(subset=['U.S. viewers (millions)'])
df = df.rename(columns={'U.S. viewers (millions)': 'viewers'})
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

def reglin_train_evaluate(df, features, target):
  df = df.dropna(subset=features)
  df = df.dropna(subset=target)
  x = df[features]
  y = df[[target]]
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

  model = LinearRegression()
  model.fit(x_train, y_train)

  y_train_pred = model.predict(x_train)
  y_test_pred = model.predict(x_test)

  print("R² treino: ", r2_score(y_train, y_train_pred))
  print("R² teste: ", r2_score(y_test, y_test_pred))

  coef_table = pd.DataFrame(list(x_train.columns)).copy()
  coef_table.insert(len(coef_table.columns),"Coefs",model.coef_.transpose())
  print(coef_table)

features = ['_800','_500','_815','_700','_1000', '_other', 'season','fred','jhon','Brian','Erik','Greenblatt','Kelly','Caleb','outros']
target = 'viewers'
reglin_train_evaluate(df, features, target)
```

```python

def mlp_train_evaluate(df, features, target):
  df = df.dropna(subset=features)
  df = df.dropna(subset=target)
  x = df[features]
  y = df[[target]]
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

  ss = StandardScaler()
  x_train_scaled = ss.fit_transform(x_train)
  x_test_scaled = ss.transform(x_test)

  nn = MLPRegressor(max_iter=500)
  nn.fit(x_train_scaled, y_train.values.ravel())

  y_test_pred_nn = nn.predict(x_test_scaled)
  y_train_pred_nn = nn.predict(x_train_scaled)

  print(r2_score(y_train, y_train_pred_nn))
  print(r2_score(y_test, y_test_pred_nn))


features = ['_800','_500','_815','_700','_1000', '_other', 'season','fred','jhon','Brian','Erik','Greenblatt','Kelly','Caleb','outros']
target = 'viewers'
mlp_train_evaluate(df, features, target)

```

```python
from warnings import simplefilter
from sklearn.utils._testing import ignore_warnings
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)

def mlp_train_evaluate_grid_search(df, features, target):
  df = df.dropna(subset=features)
  df = df.dropna(subset=target)
  x = df[features]
  y = df[[target]]
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

  ss = StandardScaler()
  x_train_scaled = ss.fit_transform(x_train)
  x_test_scaled = ss.transform(x_test)

  # optimizing
  parameters = {
    'hidden_layer_sizes': [(5,), (10,), (10, 5,), (20, 10,)],
    'activation': ['relu', 'logistic', 'tanh'],
    'solver': ['adam', 'lbfgs'],
    'learning_rate': ['constant', 'adaptive'],
  }
  nn = MLPRegressor(max_iter=500)
  clf = GridSearchCV(nn, parameters, refit=True, return_train_score=True)
  clf.fit(x_train_scaled, y_train.values.ravel())

  print("Melhores parâmetros:")
  print(clf.best_params_)

  nn_optimized = clf.best_estimator_

  y_test_pred_nn = nn_optimized.predict(x_test_scaled)
  y_train_pred_nn = nn_optimized.predict(x_train_scaled)

  print(r2_score(y_train, y_train_pred_nn))
  print(r2_score(y_test, y_test_pred_nn))


features = ['_800','_500','_815','_700','_1000', '_other', 'season','fred','jhon','Brian','Erik','Greenblatt','Kelly','Caleb','outros']
target = 'viewers'
mlp_train_evaluate_grid_search(df, features, target)
```
