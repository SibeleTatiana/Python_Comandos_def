```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv('/content/cnpjs.csv', sep=";")

# Concatenar as três primeiras colunas em uma única coluna 'cnpj'
df['cnpj'] = df['CNPJ BÁSICO'].astype(str) + df['CNPJ ORDEM'].astype(str) + df['CNPJ DV'].astype(str)

# Selecionar as colunas 'DATA DE INÍCIO ATIVIDADE' e 'CNAE FISCAL PRINCIPAL'
# e renomeá-las para 'data_abertura' e 'codigo_atividade_principal'
df = df.rename(columns={'DATA DE INÍCIO ATIVIDADE': 'data_abertura', 'CNAE FISCAL PRINCIPAL': 'codigo_atividade_principal'})

# Manter apenas as colunas desejadas
df = df[['cnpj', 'data_abertura', 'codigo_atividade_principal']]

# Exibir o DataFrame resultante
print(df)
```
