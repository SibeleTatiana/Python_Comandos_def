```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('/home/tips.csv')
df.head(2)
```

```python
print(df['time'].value_counts())
print(df['day'].value_counts())
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
```

```python
df['sex'] = df.apply(lambda row: 1 if row['sex'] == 'Female' else 0, axis=1)
df['smoker'] = df.apply(lambda row: 1 if row['smoker'] == 'No' else 0, axis=1)
df['day'] = df.apply(lambda row: 1 if row['day'] == 'Sun' else 0, axis=1)
df['time'] = df.apply(lambda row: 1 if row['time'] == 'Dinner' else 0, axis=1)
df.head(5)
```

```python
df = df[df['total_bill'] > 0]
df = df[df['tip'] > 0]
df['percgorjeta'] = df['tip'] / df['total_bill']
median = df['percgorjeta'].median()
df['tipo_gorjeta'] = df.apply(lambda row: 1 if row['percgorjeta'] > median else 0, axis=1)
df.head(5)
```

```python
features = ['sex','smoker','day','time','size']
target = 'tipo_gorjeta'
x_train, x_test, y_train, y_test = train_test_split(df[features],df[target], test_size = 0.8, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

accuracies = {}
for i in range(1,11):
  clf = DecisionTreeClassifier(max_depth=i, random_state=42)
  clf.fit(x_train, y_train)
  y_pred = clf.predict(x_test)
  accuracies[i] = metrics.accuracy_score(y_test, y_pred)
print(accuracies)
```

```python
accuracies = {}
for i in [10,100,200,400,600,800]:
  clf = RandomForestClassifier(max_depth=4, n_estimators=i, random_state=42)
  clf.fit(x_train, y_train)
  y_pred = clf.predict(x_test)
  accuracies[i] = metrics.accuracy_score(y_test, y_pred)
print(accuracies)


```
