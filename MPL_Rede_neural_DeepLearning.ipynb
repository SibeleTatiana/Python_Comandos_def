```python
import pandas as pd
df = pd.read_csv('/home/gym_members_exercise_tracking.csv')
df.head(5)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.neural_network import MLPRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

print(df.shape)
df.head(5)

df['Workout_Type'].value_counts()
```

```python
df1.dtypes
```

```python
df1 = df.copy()
df1['class1'] = df.apply( lambda row: 1 if row['Workout_Type']=='Yoga' else 0 , axis=1)
df1['class2'] = df.apply( lambda row: 1 if row['Workout_Type']=='HIIT' else 0 , axis=1)
df1['class3'] = df.apply( lambda row: 1 if row['Workout_Type']=='Strength' else 0 , axis=1)
df1['class4'] = df.apply( lambda row: 1 if row['Workout_Type']=='Other' else 0 , axis=1)
df1['Gender'] = df.apply( lambda row: 1 if row['Gender']=='Male' else 0 , axis=1)

del df1['Workout_Type']
df1.head(5)
```

```python
print(len(df1))
print(df1['Experience_Level'].value_counts())
```

```python
df1.dtypes
```

```python
df1_1 = df1[df1['Experience_Level']==1].sample(3000, replace=True, random_state=1)
df1_2 = df1[df1['Experience_Level']==2].sample(3000, replace=True, random_state=1)
df1_3 = df1[df1['Experience_Level']==3].sample(3000, replace=True, random_state=1)
df2 = pd.concat([df1_1, df1_2, df1_3])
```

```python
x = df2[['Age',
         'Gender',
 'Weight (kg)',
 'Height (m)',
 'Max_BPM',
 'Avg_BPM',
 'Resting_BPM',
 'Session_Duration (hours)',
 'Calories_Burned',
 'Fat_Percentage',
 'Water_Intake (liters)',
 'Workout_Frequency (days/week)',
 'BMI',
 'class1',
 'class2',
 'class3',
 'class4']]
y = df2[['Experience_Level']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.50, random_state=42)

ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

nn = MLPClassifier(random_state=42,max_iter=1500,
                          hidden_layer_sizes=(20,20,20),
                          activation = 'relu',
                          alpha = 0.005,
                          solver = 'adam')
nn.fit(x_train_scaled, y_train.values.ravel())

y_test_pred_nn = nn.predict(x_test_scaled)
y_train_pred_nn = nn.predict(x_train_scaled)

print(r2_score(y_train, y_train_pred_nn))
print(r2_score(y_test, y_test_pred_nn))
```

```python
x = df1[['Age',
 'Gender',
 'Weight (kg)',
 'Height (m)',
 'Max_BPM',
 'Resting_BPM',
 'Session_Duration (hours)',
 'Calories_Burned',
 'Fat_Percentage',
 'Water_Intake (liters)',
 'Workout_Frequency (days/week)',
 'BMI',
 'Experience_Level',
 'class1',
 'class2',
 'class3',
 'class4']]
y = df1[['Avg_BPM']] # TENHO QUE USAR UM REGRESSOR!!!!!

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=42)

ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

model = MLPRegressor(random_state=42, max_iter=1900, hidden_layer_sizes=(110,100,110), activation='relu', solver='adam',alpha=0.001)
model.fit(x_train_scaled, y_train)

y_train_pred = model.predict(x_train_scaled)
y_test_pred = model.predict(x_test_scaled)

print(r2_score(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred))
```
